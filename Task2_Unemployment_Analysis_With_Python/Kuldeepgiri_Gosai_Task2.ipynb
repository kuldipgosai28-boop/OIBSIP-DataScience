```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv(r"C:\Users\kuldi\Downloads\Unemployment in India.csv")

data.head()
```

```python
data.info()
```

```python
print(data.columns)
```

```python
data.columns = data.columns.str.strip()

print(data.columns)
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x="Region",
    y="Estimated Unemployment Rate (%)",
    data=data
)

plt.xticks(rotation=90)

plt.show()
```

```python
avg = data["Estimated Unemployment Rate (%)"].mean()

print("Average Unemployment Rate:", avg)
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(data.corr(numeric_only=True), annot=True)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(
    x=data.index,
    y=data["Estimated Unemployment Rate (%)"]
)

plt.show()
```
